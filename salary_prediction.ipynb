```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
```

```python
sal_data1['Gender_Encode']=Label_Encoder.fit_transform(sal_data1['Gender'])
sal_data1['Degree_Encode']=Label_Encoder.fit_transform(sal_data1['Degree'])
sal_data1['Job_Encode']=Label_Encoder.fit_transform(sal_data1['Job_Title'])

sal_data1.head(10)
```

feature selection

```python
ind=sal_data1[['Job_Encode','Gender_Encode','Degree_Encode','Experience_years']]
dep=sal_data1[['Salary']]

```

```python
ind.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
ind_train,ind_test,dep_train,dep_test=train_test_split(ind,dep,test_size=0.2,random_state=42)
dep_train
```

```python
from sklearn.linear_model import LinearRegression
Linear_model=LinearRegression()
```

```python
Linear_model.fit(ind_train,dep_train)
```

```python
test=Linear_model.predict(ind_test)
test
```

```python
df=dep_test[['Salary']]
df['predicted']=test
df
```

```python
plt.scatter(df['Salary'],df['predicted'],color='Green',marker="o",s=100)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.show()
```

```python
print(len(ind_train))
print(len(dep_train))
```

```python


sns.distplot(test)
```

```python
df['error']=df['Salary']-df['predicted']
df['abs_error']=abs(df['error'])
mean_error=df['abs_error'].mean()
mean_error
df['squared_error'] = df['error'] ** 2
mse = df['squared_error'].mean()
rmse = np.sqrt(mse)
percentage_error = (df['error'] / dep_test['Salary']) * 100
absolute_percentage_error = abs(percentage_error)
mape = absolute_percentage_error.mean()
mpe = percentage_error.mean()
mean_error
mse
```

```python
from sklearn.metrics import r2_score
```

```python

print("Accuracy of the model is:",round(r2_score(dep_test,test)*100),"%")
```

```python
import tkinter as tk
from tkinter import scrolledtext


# Create the GUI window
root = tk.Tk()
root['bg']='cyan'
root.title("Salary Prediction")
root.geometry("800x600")  # Set window size to 800x600 pixels

# Create a Text widget to display new_enc.head(60)
text_area = scrolledtext.ScrolledText(root, width=100, height=20)
text_area.pack()

# Function to predict salary
def predict_salary():
    try:
        # Get the values from the entry fields
        job_encode = int(job_entry.get())
        gender_encode = int(gender_entry.get())
        degree_encode = int(degree_entry.get())
        experience_years = int(experience_entry.get())
        
        # Predict the salary using the linear model
        emp_salary = new.predict([[job_encode, gender_encode, degree_encode, experience_years]])
        predicted_salary_label.config(text="\n Predicted Salary: RS" + str(round(emp_salary[0])))
    except Exception as e:
        predicted_salary_label.config(text="Error predicting salary: " + str(e))

# Create entry fields for user input
job_label = tk.Label(root, text="Job Encode:")
job_label.pack()
job_entry = tk.Entry(root)
job_entry.pack()

gender_label = tk.Label(root, text="Gender Encode: (0:FEMALE ,1:MALE)")
gender_label.pack()
gender_entry = tk.Entry(root)
gender_entry.pack()

degree_label = tk.Label(root, text="Degree Encode: 0:Bachelor's,1:Master,2:Phd")
degree_label.pack()
degree_entry = tk.Entry(root)
degree_entry.pack()

experience_label = tk.Label(root, text="Experience (years):")
experience_label.pack()
experience_entry = tk.Entry(root)
experience_entry.pack()

predict_button = tk.Button(root, text="Predict Salary", command=predict_salary)
predict_button.pack()

predicted_salary_label = tk.Label(root, text="Predicted Salary: ")
predicted_salary_label.pack()

# Display new_enc.head(60) in the text widget
text_area.delete("1.0", tk.END)  # Clear the text widget
text_area.insert(tk.END, str(new_enc.head(60)))

root.mainloop()
```
